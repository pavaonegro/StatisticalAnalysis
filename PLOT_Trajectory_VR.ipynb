# **DATA FORMAT CONVERSION - FROM String of Coordinates TO List of Tuples**

```python
def convert_data_to_tuples(data):
  """Converts a string of coordinate pairs into a list of tuples.

  Args:
    data: A string containing coordinate pairs separated by newlines.

  Returns:
    A list of tuples, where each tuple represents a coordinate pair.
  """

  lines = data.strip().split('\n')
  coordinates = []
  for line in lines:
    x, y = map(float, line.split(','))
    coordinates.append((x, y))
  return coordinates

# Example usage:
data = """
5014.981284,-423.420257
5024.881306,-582.046447
4897.444829,-195.261127
4824.86315,9.792558
4860.424425,197.737997


"""

coordinates = convert_data_to_tuples(data)
print(coordinates)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_points_with_gradient(start_point, points):
    """Plots a set of points in sequence with a color gradient from blue to red.

    Args:
        start_point: The starting point as a tuple (x, y).
        points: A list of tuples representing the points in sequence.
    """

    # Combine start point and points into a single list
    all_points = [start_point] + points

    # Filter out consecutive duplicate points
    filtered_points = [all_points[i] for i in range(len(all_points)) if i == 0 or all_points[i] != all_points[i - 1]]

    # Convert the list of tuples into NumPy arrays for x and y coordinates
    x_coords, y_coords = np.array(filtered_points).T

    # Create a larger plot for better visibility
    plt.figure(figsize=(12, 8))

    # Number of points
    num_points = len(x_coords)

    # Create a colormap
    cmap = plt.get_cmap('plasma')

    # Normalize the color range to the number of connections
    colors = [cmap(i / (num_points - 1)) for i in range(num_points - 1)]

    # Plot each segment between consecutive points with a color gradient
    for i in range(num_points - 1):
        plt.plot(x_coords[i:i + 2], y_coords[i:i + 2], color=colors[i], marker='o')
        plt.scatter(x_coords[i], y_coords[i], color=colors[i], s=50, zorder=5)  # Highlight each point

    # Highlight the start and end points for clarity
    plt.scatter(x_coords[0], y_coords[0], color='green', s=100, label='Start', zorder=10)
    plt.scatter(x_coords[-1], y_coords[-1], color='red', s=100, label='End', zorder=10)

    plt.grid(True)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Plot of Points with Gradient Color')
    plt.legend()
    plt.show()

# Define your starting point and relative points
start_point = (4573.049868, -527.320728)
relative_points = [(5014.981284, -423.420257), (5024.881306, -582.046447), (4897.444829, -195.261127), (4824.86315, 9.792558), (4860.424425, 197.737997), (4763.790653, 370.821306), (4749.436343, 559.062146), (4764.08743, 780.678336), (4762.478093, 906.078522), (4754.397986, 1009.327129), (4796.523671, 1004.413228), (4698.387209, 1065.27263), (4777.385095, 1065.356436), (4724.999283, 1043.837468), (5018.232301, 859.945704), (5193.128271, 908.661451), (5095.622773, 958.968752), (5146.488068, 1032.062281), (5208.41561, 974.7866), (5218.581428, 863.943844), (5166.697895, 923.164695), (5157.286693, 977.197531), (5186.268548, 881.079729), (5183.176887, 931.090137), (5239.568599, 932.845393), (5211.200264, 986.940686), (5159.048451, 951.448245), (5096.644551, 929.302624), (5127.392974, 957.494921), (5150.032101, 923.508804), (5019.841838, 925.592219), (4831.728827, 928.351895), (4629.463944, 952.746264), (4367.815759, 967.261047), (4229.21845, 974.791876), (4073.889406, 996.494677), (4191.772675, 997.83358), (4093.188022, 1017.612245), (4191.889767, 1029.495521), (4124.846952, 1047.434002), (4159.821693, 1004.518501), (4190.119074, 968.555013), (4412.529958, 996.673948), (4631.473359, 997.107329), (4857.090806, 974.409358)]

plot_points_with_gradient(start_point, relative_points)
```

# **HMD ROTATION**

```python
def convert_data_format(data_string):
  """Converts data from HMD Rotation format to space-separated values.

  Args:
    data_string: The input data string.

  Returns:
    A list of converted lines.
  """

  lines = data_string.splitlines()
  converted_data = []
  for line in lines:
    line = line.replace("HMD Rotation: ", "").replace("P=", "").replace("Y=", "").replace("R=", "")
    converted_data.append(line)
  return converted_data

# Sample data
data_string = """
HMD Rotation: P=-14.239892 Y=21.198532 R=-0.156497
HMD Rotation: P=-18.369587 Y=4.081471 R=-2.506471
HMD Rotation: P=-30.961469 Y=9.279979 R=-17.758552
HMD Rotation: P=-17.556348 Y=44.750397 R=-2.241889
HMD Rotation: P=-3.538492 Y=9.727543 R=5.965961

"""

converted_data = convert_data_format(data_string)

# Print the converted data
for line in converted_data:
  print(line)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.cm import get_cmap  # For compatibility with older Matplotlib versions

def parse_data(data_string):
  """Parses the data string into a list of tuples."""
  lines = data_string.splitlines()
  data = []
  value_counts = {}  # Dictionary to store value counts

  for line in lines:
    line = line.strip()  # Remove leading and trailing whitespace
    if not line:  # Skip empty lines
      continue
    try:
      values = line.split()
      p, y, r = map(float, values)
      data.append((p, y, r))

      # Update value counts
      key = (p, y, r)  # Use tuple as unique identifier
      value_counts[key] = value_counts.get(key, 0) + 1
    except ValueError:
      print(f"Error parsing line: {line}")  # Handle invalid lines

  return data, value_counts

def plot_3d_scatter(data, value_counts):
  """Plots a 3D scatter plot of the data with color based on heatmap."""
  x, y, z = zip(*data)

  # Normalize value counts for color mapping
  max_count = max(value_counts.values())
  colors = get_cmap('Reds')(np.array([value_counts[key] / max_count for key in data]))

  fig = plt.figure(figsize=(12, 8))
  ax = fig.add_subplot(111, projection='3d')
  # Create a scatter object (mappable) for the colorbar
  p = ax.scatter(x, y, z, c=colors)
  ax.set_xlabel('Pitch')
  ax.set_ylabel('Yaw')
  ax.set_zlabel('Roll')
  plt.colorbar(p, label='Frequency (Normalized)')  # Use the scatter object for colorbar
  plt.show()

# Example usage:
data_string = """

-14.239892 21.198532 -0.156497
-18.369587 4.081471 -2.506471
-30.961469 9.279979 -17.758552
-17.556348 44.750397 -2.241889
-3.538492 9.727543 5.965961


"""

data, value_counts = parse_data(data_string)
plot_3d_scatter(data, value_counts)

```

```python
# ============================================
#  Boxplot of Time per Condition
# ============================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, ttest_rel, mannwhitneyu

# -----------------------------------
#  Data
# -----------------------------------
# Each array is the time (seconds) each user took in the condition
time_A = np.array([49.08425, 57.7995, 43.6975, 111.21725, 34.117, 50.93175, 93.96575, 60.94425, 34.512, 86.1395, 57.7145, 52.09175, 80.79875, 51.55925, 120.96275, 39.47, 34.78175, 25.56125, 122.07, 66.5795, 117.59625, 84.32225, 53.556, 52.39575, 145.04425, 50.84725, 26.585, 41.275, 74.51825])
time_B = np.array([75.2075, 53.2295, 24.0305, 39.1735, 102.047, 24.37, 110.049, 79.4513, 36.6293, 72.3387, 84.407, 41.5688, 158.395, 103.792, 101.549, 75.9978, 35.5313, 111.165, 60.033, 127.577, 65.5453, 139.233, 68.0452, 109.95, 47.9568, 53.8468, 27.5768, 107.905, 90.0907])

# -----------------------------------
# Statistical tests
# -----------------------------------
shapiro_A = shapiro(time_A)
shapiro_B = shapiro(time_B)

print(f"Shapiro-Wilk Test A: p={shapiro_A.pvalue:.4f}")
print(f"Shapiro-Wilk Test B: p={shapiro_B.pvalue:.4f}")

if shapiro_A.pvalue > 0.05 and shapiro_B.pvalue > 0.05:
    stat, p = ttest_rel(time_A, time_B)
    test_name = "Paired t-test (parametric)"
else:
    stat, p = mannwhitneyu(time_A, time_B, alternative='two-sided')
    test_name = "Mann-Whitney U test (non-parametric)"

print(f"\n {test_name}")
print(f"Statistic = {stat:.4f}, p-value = {p:.4f}")
if p < 0.05:
    print("Significant difference between A and B.")
else:
    print("No significant difference between A and B.")

# -----------------------------------
# Boxplot visualization
# -----------------------------------
fig, ax = plt.subplots(figsize=(7, 6))

# Boxplot settings
box = ax.boxplot([time_A, time_B],
                 patch_artist=True,
                 labels=['Condition A', 'Condition B'],
                 widths=0.5,
                 medianprops=dict(color='black', linewidth=1.5),
                 boxprops=dict(linewidth=1.5, color='black'),
                 whiskerprops=dict(linewidth=1.5, color='black'),
                 capprops=dict(linewidth=1.5, color='black'),
                 showmeans=True,
                 meanprops=dict(marker='o', markerfacecolor='white', markeredgecolor='black', markersize=8))

# Custom colors
colors = ['#69b3ff','#ff7f7f']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

# Overlay individual points
x_positions = [1, 2]
for i, times in enumerate([time_A, time_B]):
    x_jitter = np.random.normal(loc=x_positions[i], scale=0.03, size=len(times))
    plt.scatter(x_jitter, times, color='black', s=50, alpha=0.5, zorder=3)

# Add labels and title
ax.set_title('Task Time', fontsize=14, pad=15)
ax.set_ylabel('Time (seconds)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.5, axis='y')

# # Add mean  SD annotation above boxes
# means = [np.mean(time_A), np.mean(time_B)]
# stds = [np.std(time_A, ddof=1), np.std(time_B, ddof=1)]
# for i, (m, s) in enumerate(zip(means, stds), start=1):
#     ax.text(i, m + s + 1.5, f"{m:.1f}  {s:.1f}", ha='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

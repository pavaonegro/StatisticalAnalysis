```python
import numpy as np
from scipy import stats

def normalize_and_remove_outliers(arr, method='z-score', threshold=3, norm_range=(0, 1)):
    """
    Normalize the array and remove outliers.

    Parameters:
    - arr: List or NumPy array of numbers.
    - method: Method for outlier detection ('z-score' or 'iqr').
    - threshold: Threshold for outlier detection.
    - norm_range: Tuple indicating the range for normalization (default is (0, 1)).

    Returns:
    - normalized_arr: Cleaned and normalized NumPy array.
    - removed_count: Number of data points removed as outliers.
    - removed_percentage: Percentage of data points removed as outliers.
    - removed_indices: List of indices of the removed data points.
    - removed_values: List of values of the removed data points.
    """

    arr = np.array(arr)
    original_size = arr.size

    if original_size == 0:
        raise ValueError("Input array is empty.")

    # Initialize lists for removed indices and values
    removed_indices = []
    removed_values = []

    # Remove outliers
    if method == 'z-score':
        z_scores = np.abs(stats.zscore(arr))
        outliers_mask = z_scores >= threshold
    elif method == 'iqr':
        Q1 = np.percentile(arr, 25)
        Q3 = np.percentile(arr, 75)
        IQR = Q3 - Q1
        outliers_mask = (arr < (Q1 - 1.5 * IQR)) | (arr > (Q3 + 1.5 * IQR))
    else:
        raise ValueError("Method should be 'z-score' or 'iqr'")

    removed_indices = np.where(outliers_mask)[0].tolist()
    removed_values = arr[outliers_mask].tolist()
    filtered_arr = arr[~outliers_mask]

    # Calculate the number and percentage of data points removed
    removed_count = original_size - filtered_arr.size
    removed_percentage = (removed_count / original_size) * 100

    # Check if filtering left any data
    if filtered_arr.size == 0:
        raise ValueError("All data points were removed as outliers.")

    # Normalize the filtered array
    min_val = np.min(filtered_arr)
    max_val = np.max(filtered_arr)
    if min_val == max_val:
        normalized_arr = np.full(filtered_arr.shape, norm_range[0])  # Handle edge case where all values are identical
    else:
        norm_min, norm_max = norm_range
        normalized_arr = (filtered_arr - min_val) / (max_val - min_val)
        normalized_arr = normalized_arr * (norm_max - norm_min) + norm_min

    return normalized_arr, removed_count, removed_percentage, removed_indices, removed_values

# Example usage
data = [26.12325, 4.57, 19.667, 72.04375, 67.93, 26.56175, 16.08325, 18.50705, 2.1173, 307.13525, 13.8008, 26.6925, 10.52295, 77.59625, 52.23275, 19.41375, 36.5278, 0.74955, 85.60375, 62.037, 60.9975, 52.05095, 54.91075, 135.39055, 14.4892, 57.55425, 97.08745, 2.99955, 0.9918, 66.63, 15.57245]
normalized_data, removed_count, removed_percentage, removed_indices, removed_values = normalize_and_remove_outliers(data, method='iqr')

print(f"Normalized Data: {normalized_data}")
print(f"Number of data points removed: {removed_count}")
print(f"Percentage of data points removed: {removed_percentage:.2f}%")
print(f"Removed Indices: {removed_indices}")
print(f"Removed Values: {removed_values}")

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

def check_normality(data):
    """
    Check if the data follows a normal distribution using visual and statistical methods.

    Parameters:
    - data: List or NumPy array of numbers.

    Outputs:
    - Histogram with KDE and Q-Q plot.
    - Shapiro-Wilk test result.
    - Kolmogorov-Smirnov test result.
    """

    # Convert data to a NumPy array if it's not already
    data = np.array(data)

    # Plotting histogram with Kernel Density Estimate (KDE)
    plt.figure(figsize=(14, 6))

    plt.subplot(1, 2, 1)
    sns.histplot(data, kde=True)
    plt.title("Histogram with KDE")
    plt.xlabel("Data")
    plt.ylabel("Frequency")

    # Q-Q plot
    plt.subplot(1, 2, 2)
    stats.probplot(data, dist="norm", plot=plt)
    plt.title("Time Data Distribution")

    plt.tight_layout()
    plt.show()

    # Shapiro-Wilk Test
    shapiro_stat, shapiro_p_value = stats.shapiro(data)
    print(f"Shapiro-Wilk Test: W={shapiro_stat}, p-value={shapiro_p_value}")

    # Kolmogorov-Smirnov Test
    ks_stat, ks_p_value = stats.kstest(data, 'norm', args=(np.mean(data), np.std(data)))
    print(f"Kolmogorov-Smirnov Test: KS-statistic={ks_stat}, p-value={ks_p_value}")

    # Interpretation of results
    alpha = 0.05  # Common significance level
    if shapiro_p_value < alpha:
        print("Shapiro-Wilk Test: Reject the null hypothesis (data is not normally distributed).")
    else:
        print("Shapiro-Wilk Test: Fail to reject the null hypothesis (data is normally distributed).")

    if ks_p_value < alpha:
        print("Kolmogorov-Smirnov Test: Reject the null hypothesis (data is not normally distributed).")
    else:
        print("Kolmogorov-Smirnov Test: Fail to reject the null hypothesis (data is normally distributed).")

# Example usage
data = np.array([48.2505, 65.662, 44.7755, 114.301, 32.5125, 54.889, 84.233, 64.6685, 28.429, 56.845, 53.8515, 46.8495, 75.767, 25.937, 89.637, 39.762, 31.611, 20.403, 64, 51.6935, 119.2945, 97.2665, 48.2625, 58.667, 125.3755, 53.2525, 21.462, 35.1245, 48.7135, 54.0875, 54.569, 21.016, 41.4125, 46.24, 24.8985, 80.9675, 66.1855, 37.075, 72.8955, 73.8575, 37.367, 161.531, 90.319, 84.3405, 76.9765, 31.757, 91.907, 57.45, 145.307, 45.241, 138.16, 71.109, 98.158, 50.1725, 48.901, 23.1605, 64.3055, 57.2135])  # Replace with your actual data
check_normality(data)
```
